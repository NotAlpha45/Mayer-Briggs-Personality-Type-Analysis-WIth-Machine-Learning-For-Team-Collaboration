```python
import re
import pandas as pd
```

# Dataset Cleaning
The dataset contains explicit mentions of personality type in the text, which is not desireable for the training. Hence, we will omit it.

```python
words_to_remove = [
    "ENTP",
    "ENTJ",
    "INTP",
    "INTJ",
    "INFJ",
    "INFP",
    "ENFJ",
    "ENFP",
    "ISTJ",
    "ISFJ",
    "ESTJ",
    "ESFJ",
    "ISTP",
    "ISFP",
    "ESTP",
    "ESFP",
]

placeholder = "[PERSONALITY]"
```

```python
dataframe = pd.read_parquet(
    "../datasets/label_seperated_MBTI_500.pq", engine="fastparquet"
)
dataframe
```

```python
omit_pattern = re.compile(
    r"\b(?:{})\b".format("|".join(map(re.escape, words_to_remove))), re.IGNORECASE
)
```

```python
dataframe["posts"] = dataframe["posts"].apply(lambda text: omit_pattern.sub(placeholder, text))
```

```python
dataframe.to_parquet("../datasets/label_seperated_cleaned_MBTI_500.pq")
```

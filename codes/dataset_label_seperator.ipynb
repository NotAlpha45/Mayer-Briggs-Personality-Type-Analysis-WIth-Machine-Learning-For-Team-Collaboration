### Feature Seperation 
Using ```str.split("", expand=True)``` we seperated each of the letters in the personality type string and made them into seperate columns. By selecting the indices 1, 2, 3 and 4 we took only the columns containing letters, not the empty strings. Later on, we assigned the columns to the dataframe.   

````python
feature_seperated_columns = test_dataframe["type"].str.split("", expand=True)[[1, 2, 3, 4]]
feature_seperated_columns
````

````python
test_dataframe[["I_E", "N_S", "T_F", "J_P"]] = feature_seperated_columns
````
